```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("tips")
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.histplot(df["total_bill"])
plt.title("Distribution of Total Bill")
plt.show()
```

```python
sns.barplot(x="day", y="total_bill", data=df)
plt.title("Average Total Bill by Day")
plt.show()
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
print("EDA Conclusion:")
print("1. Total bill and tip have a strong positive correlation.")
print("2. Larger group size generally leads to higher total bills.")
print("3. Weekend days (Saturday and Sunday) show higher bills.")
print("4. No missing values were found in the dataset.")
```
